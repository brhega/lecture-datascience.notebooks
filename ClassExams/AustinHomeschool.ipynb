The contents of the csv is year vs number of crimes that I took from question 1 of the test, then added another column for the number of withdrawals for the county of Travis TX (Austin)

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV file into a DataFrame
df = pd.read_csv('AustinCrimeVSHomeschool.csv')
```

```python
chart = df.plot(x='Year', y=['Crimes', 'Withdrawals'], kind='bar', figsize=(10,6))
chart.set_ylabel('Counts')
chart.set_title('Crimes vs Withdrawals in Austin, TX by Year')
```

with withdrawals being a much smaller number this isn't super helpful.  Maybe a dataframe with percent changes could be better.  I'm thinking this because it could be a good way to see if a current year increase in crime lead to increased withdrawals in the same year. 

```python
df_percChanges = df.copy()

df_percChanges['Crimes_Percent_Change'] = df['Crimes'].pct_change() * 100
df_percChanges['Withdrawals_Percent_Change'] = df['Withdrawals'].pct_change() * 100

df_percChanges
```

this is better to analyze, though any correlation isn't being made super obvious
